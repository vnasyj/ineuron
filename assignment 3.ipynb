```python
#Write a Python Program to Check if a Number is Positive, Negative or Zero?
num=-10
if(num>0):
    print("it's a positive number",num)
elif(num==0):
    print("it's a zero number",num)
else:
    print("it's negative number",num)
```

```python
#Write a Python Program to Check if a Number is Odd or Even?
num=13
if(num%2==0):
    print("it's even number",num)
else:
    print("it's odd number",num)
```

```python
#Write a Python Program to Check Leap Year?
a=2022
if(a%4==0 and a%100==0):
    print("it's a leap year",a)
else:
    print("it's not leap year",a)
```

```python
# Write a Python Program to Check Prime Number
num1=11
if(num1%2==0):
    print("it's not prime number",num1)
else:
    print("it's prime number",num1)
```
